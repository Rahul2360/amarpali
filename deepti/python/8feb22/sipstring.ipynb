```python
def swap(charac, i, j):
    temp = charac[i]
    charac[i] = charac[j]
    charac[j] = temp
def permutations(charac, firstindex=0):
    arr = []
    if firstindex == len(charac) - 1:
        b = ''.join(charac)
        arr.append(b)
        print(arr)
 
    for i in range(firstindex, len(charac)):
        swap(charac, firstindex, i)
        permutations(charac, firstindex + 1)
        swap(charac, firstindex, i)
 
s = 'depti'
permutations(list(s))
    
```

```python
investedmonthly_amount=int(input("Enter Invested amount: "))
time=int(input("Enter time in years : "))
rate=int(input("Enter rate of return : "))

def sip_calculator(monthly_sip_amount,time,rate):
    monthly_rate=rate/12/100
    months=time*12
    gained = (monthly_sip_amount*(((1+monthly_rate)*months)-1)*(1+monthly_rate))/monthly_rate
    invested_amount=monthly_sip_amount*months
    rounded_gained = round(gained)
    print("Invested Amount : ",invested_amount)
    print("Resultant Amount : ",rounded_gained)
    
sip_calculator(investedmonthly_amount,time,rate)
```

```python
p = float(input('Enter the principle amount: '))
t = float(input('Enter the time: '))
r = float(input('Enter the rate: '))
def compound_interest(p, r, t):
 
    Amount = p * (pow((1 + r / 100), t))
    CI = Amount - p
    print("Compound interest is", CI)
 

compound_interest(p,r,t)
```
