```python
import xlsxwriter
import random
import names
import openpyxl as xl
from openpyxl import Workbook

workbook = xlsxwriter.Workbook('/home/deepti/Downloads/mydata_file.xlsx')
worksheet = workbook.add_worksheet()
worksheet.write('A1', 'Id')
worksheet.write('B1', 'Names')
worksheet.write('C1', 'Age')
worksheet.write('D1', 'Gender')
rowIndex = 2
for row in range(100):
    Id = row+1
    Names = names.get_first_name()
    Age = 20 + random.randint(1, 100)
    Gender = random.choice(['Female','Male'])
    
    worksheet.write('A' + str(rowIndex), Id)
    worksheet.write('B' + str(rowIndex), Names)
    worksheet.write('C' + str(rowIndex), Age)
    worksheet.write('D' + str(rowIndex), Gender)
    
    rowIndex += 1
    print(Id, Names, Age, Gender)

workbook.close()
```
