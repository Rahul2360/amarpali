```python
import openpyxl as xl
from openpyxl import Workbook
import json
wb_s=xl.load_workbook("T90_POL.xlsx")
user_data=input("enter policy number  ")
final_list=[]
final_dict={}
ws_names=wb_s.sheetnames
for i in range(0,93):
    ws_s=wb_s.worksheets[i]
    current_list=[]
    rows=ws_s.max_row
    cols=ws_s.max_column
    for j in range(2,rows+2):
        new_dict={}
        vals=ws_s.cell(row=j,column=1).value
        if vals==user_data:
            for k in range(1,cols+1):
                        temp=(ws_s.cell(row=1,column=k).value)    
                        new_dict[temp]= ws_s.cell(row=j,column=k).value
#                         print(k)
            current_list.append(new_dict)
            final_dict[ws_names[i]]=current_list
                #break if we want to print repeated data  regarding repeated policy no. only once 
                #also remove current_list==[] condition for simplicity only
        elif j==rows+1 and len(current_list)==0: 
            current_list.append(new_dict)
            final_dict[ws_names[i]]=current_list
# print(type(user_data))
# print(type(vals))
final_list.append(final_dict)
print("Total Sheets= "+str(len(final_dict)))
final_data=json.dumps(final_list,sort_keys=False,indent=3)
print(final_data)
```
