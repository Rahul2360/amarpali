```python
import openpyxl as xl
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.styles import PatternFill


source ="/home/devendra/Downloads/short_term_times.xlsx"
workbook_s = xl.load_workbook(source)
worksheet_s = workbook_s.worksheets[0]

workbook = Workbook()
worksheet =  workbook.active
worksheet.title = "Changed Sheet"
workbook.save(filename = 'my_file.xlsx')

filename1 ="/home/devendra/notebook/my_file.xlsx"
workbook_d = xl.load_workbook(filename1)
worksheet_d = workbook_d.active
worksheet_d['E1']="PRIORITY"
worksheet_d['F1']="SEVERITY"

   

row_s = worksheet_s.max_row

for i in range (1, row_s + 1):
    for j in range (7,11):
        c = worksheet_s.cell(row = i, column = j)
        worksheet_d.cell(row = i, column = j-6).value = c.value
        
for i in range (1, row_s + 1):
    for j in range (11,13):
        
        c = worksheet_s.cell(row = i, column = j)
        worksheet_d.cell(row = i, column = j-4).value = c.value


for i in range (2, row_s + 1):
    for j in range (11,12):
        c = worksheet_s.cell(row = i, column = j)
        if c.value=="72 hours":
             worksheet_d.cell(row = i, column = j-6).value = "P3"
             worksheet_d.cell(row = i, column = j-5).value = "S3"
        elif c.value=="36 hours":
              worksheet_d.cell(row = i, column = j-6).value = "P2"
              worksheet_d.cell(row = i, column = j-5).value = "S2"
        elif c.value=="24 hours":
              worksheet_d.cell(row = i, column = j-6).value = "P1"
              worksheet_d.cell(row = i, column = j-5).value = "S1"
        else:
              worksheet_d.cell(row = i, column = j-6).value = "P4"
              worksheet_d.cell(row = i, column = j-5).value = "S4"

for j in range (1,9):
         worksheet_d.cell(row = 1, column = j).font=Font(bold="true",color="FFFFFF")
         worksheet_d.cell(row = 1, column = j).fill=PatternFill("solid" , start_color="373a3a")

workbook_d.save(str(filename1))
workbook_d.close()
workbook_s.close()
```
